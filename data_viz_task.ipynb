```python
# Import libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load data

car_data = pd.read_csv('Cars93.csv', index_col = 0) 

car_data.info()
```

### 1. Boxplot

```python
# Boxplot for revs per mile

car_data.boxplot(column="RPM", by="Manufacturer", rot=90)
plt.title("RPM")
plt.xlabel("Manufacturer")
plt.ylabel("RPM")
plt.show()



```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**


```python
# Filter the data for manufacturer 'Mazda'
mazda_data = car_data[car_data['Manufacturer'] == 'Mazda']

# Plot the boxplot for 'RPM' of Mazda vehicles
mazda_data.boxplot(column="RPM")
plt.title("RPM for Mazda Vehicles")
plt.xlabel("Mazda")
plt.ylabel("RPM")
plt.show()
```
