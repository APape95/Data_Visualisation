```python
# Import libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load data

car_data = pd.read_csv('Cars93.csv', index_col = 0) 

car_data.info()
```

### 1. Boxplot

```python
# Boxplot for revs per mile

car_data.boxplot(column="Rev.per.mile", by="Manufacturer", rot=90)
plt.title("Revs Per Mile")
plt.xlabel("Manufacturer")
plt.ylabel("Revs Per Mile")
plt.show()



```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**


```python
# Filter the data for manufacturer 'Geo'
geo_data = car_data[car_data['Manufacturer'] == 'Geo']

# Plot the boxplot for 'RPM' of Geo vehicles
geo_data.boxplot(column="Rev.per.mile")
plt.title("RPM for Geo Vehicles")
plt.xlabel("Geo")
plt.ylabel("Revs Per Mile")
plt.show()
```

### 2. Histogram

```python
# Histogram of MPG

# Create histograms for MPG.city and MPG.highway using matplotlib
plt.hist(car_data["MPG.city"], bins=10, color='skyblue', alpha=0.5, label='MPG City')
plt.hist(car_data["MPG.highway"], bins=10, color='salmon', alpha=0.5, label='MPG Highway')
plt.xlabel("Miles Per Gallon (MPG)")
plt.ylabel("Frequency")
plt.title("Histogram of MPG (City and Highway)")
plt.legend()
plt.show()
```

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  


```python
# The Turning Circle generally increases with Wheelbase increasing

# Sort data by the 'Wheelbase' column
sorted_car_data = car_data.sort_values(by='Wheelbase')

# Create a line plot for Wheelbase vs turning circle using pandas
sorted_car_data.plot(x="Wheelbase", y="Turn.circle", marker='o', linestyle='-')
plt.xlabel("Wheelbase")
plt.ylabel("Turning Circle")
plt.title("Wheelbase vs Turning Circle")
plt.grid(True)
plt.show()
```

### 4. Barplot

```python
# Barplot for mean horsepower
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

```python

# In general the larger the car the higher the horsepower

# Group the data by car size and calculate mean horsepower
mean_horsepower_by_size = car_data.groupby('Type')['Horsepower'].mean().sort_values()

# Create a bar plot
plt.figure(figsize=(10, 6))
mean_horsepower_by_size.plot(kind='bar', color='skyblue')
plt.title('Mean Horsepower by Car Size')
plt.xlabel('Car Size')
plt.ylabel('Mean Horsepower')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
